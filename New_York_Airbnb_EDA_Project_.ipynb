```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df2,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title('locality and revies dependency')
plt.show()
```

Scatter plot showing the relationship between price and number of reviews for rental listings in NYC.
Neighborhoods:
Brooklyn (blue), Manhattan (orange), Queens (green), Bronx (red), Staten Island (purple).
Price Range: Most listings are below $400; some in Manhattan reach $1400.
Review Range: Most listings have fewer than 500 reviews; some in Brooklyn and Manhattan reach 1750.
Concentration:
Majority of listings cluster at prices below $400 and reviews below 500.
Borough Insights:
Brooklyn and Manhattan: Most listings, with wide price ($0-$1400) and review (0-1750) ranges.
Queens: Moderate presence, mostly below $400 and 750 reviews.
Bronx and Staten Island: Fewer listings, prices typically under $400, reviews below 500.
Price-Review Correlation:
No strong correlation; high-priced listings (above $800) often have fewer reviews.
Market Trends:
Brooklyn and Manhattan are the most active markets.
Queens, Bronx, and Staten Island have smaller, more affordable markets.

```python
sns.pairplot(data=df2,vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'],hue='room_type')
```

The plot consists of four scatter plots and four density plots, analyzing rental listings by room type (Private room, Entire home/apt, Hotel room, Shared room) across four variables: price, minimum nights, number of reviews, and availability_365. Most listings have prices below $500, with Entire homes/apts (orange) being the most common and expensive. Minimum nights are typically under 100, with a peak around 1-10 nights. Number of reviews is mostly below 500, with a few outliers reaching 2000. Availability_365 shows a bimodal distribution, with many listings either available for 0-50 days or 300-365 days. The data highlights that most listings are affordable, short-stay, moderately reviewed, and either rarely or frequently available.

: Moderate positive correlation (0.066), suggesting listings with more beds tend to have slightly higher prices.
Moderate negative correlation (-0.12), meaning listings with longer minimum stays tend to have fewer reviews per month.
